```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sys import stdin, stdout

%matplotlib inline
```

```python
with open('fqsp.pkl', 'rb') as f:
    ans_id, ans_question, ans_body = pickle.load(f)
ans_question = np.array(ans_question)
ans_question -= 4  # --> [0, 13] (10 + oublie qqch + 3 inconnues)

# Remove invalid classes
valid_idx = ans_question <= 9
ans_question = ans_question[valid_idx]
```

```python
filenames = ['fqsp_tsne_p{}i{}.tsv'.format(*p) for p in [(20, 1000), (40, 1000)]]
data_list = []
for fname in filenames:
    d = np.genfromtxt(fname, delimiter='\t')
    d -= d.min(axis=0)
    d /= d.max(axis=0)
    data_list.append(d.copy())
```

```python
for d in data_list:
    plt.figure(figsize=(20,10))
    plt.scatter(d[:, 0], d[:, 1], c=ans_question, cmap=plt.cm.get_cmap("jet", 10))
    plt.colorbar(ticks=range(10))
    plt.show()
```
